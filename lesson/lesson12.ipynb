```python
import statistics
# 統計指標モジュール

data = [8, 17, 0, 11, 6, 21, 16, 6, 17, 11, 7, 9, 6, 13, 12, 16, 3, 14, 13, 12]
print(data)
print(sorted(data))
print(len(data))
print(f"平均値は{statistics.mean(data)}です。")
print(f"中央値は{statistics.median(data)}です。")
print(f"最頻値は{statistics.mode(data)}です。")
print(f"分散は{statistics.pvariance(data)}です。")
print(f"標準偏差は{statistics.pstdev(data)}です。")
```

```python
import math  # 数学関連機能を扱うモジュール
import matplotlib.pyplot as plt 

x = []
s = []
c = []

for i in range(50):
    x.append(i*0.05*math.pi)  # xの値を作成
    s.append(math.sin(x[i]))  # sin値を作成
    c.append(math.cos(x[i]))  # cos値を作成
    
plt.title("sin/cos functions")
plt.xlabel("rad")
plt.ylabel("value")
plt.grid(True)  # グリッドを表示

plt.plot(x, s, label="sin")  # sin()関数のグラフをラベル付きで作成
plt.plot(x, c, label="cos")  # cos()関数のグラフをラベル付きで作成
plt.legend()  # ラベルから凡例を作成

plt.show()
```

```python
import math  # 数学関連機能を扱うモジュール
import matplotlib.pyplot as plt 

x = []
s = []
c = []

for i in range(50):
    x.append(i*0.05*math.pi)  # xの値を作成
    s.append(math.sin(x[i]))  # sin値を作成
    c.append(math.cos(x[i]))  # cos値を作成
    
plt.title("sin/cos functions")
plt.xlabel("rad")
plt.ylabel("value")
plt.grid(True)  # グリッドを表示

plt.plot(x, s, label="sin", color="g")  # sin()関数のグラフをラベル付きで作成
plt.plot(x, c, label="cos", linestyle="--")  # cos()関数のグラフをラベル付きで作成
plt.legend()  # ラベルから凡例を作成

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2.5, 0.05) * np.pi
s = np.sin(x)
c = np.cos(x)

plt.title("sin/cos functions")
plt.xlabel("rad")
plt.ylabel("value")
plt.grid(True)

plt.plot(x, s, label="sin")
plt.plot(x, c, label="cos")
plt.legend()

plt.show()
```
